```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from keras import Sequential
from keras.layers import Dense,Input,Dropout
from keras.optimizers import Adam,SGD
import warnings
warnings.filterwarnings("ignore")
np.set_printoptions(precision=6, suppress=True, floatmode="fixed")
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, sep=";")
```

```python
x=wine_data.drop(['quality'],axis=1).values
x=StandardScaler().fit_transform(x)
y=wine_data['quality'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
model=Sequential()
model.add(Input(shape=x_train.shape[1]))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.3))
#model.add(Dense(32,activation='relu'))
#model.add(Dropout(0.3))
model.add(Dense(1))

model.compile(optimizer=SGD(learning_rate=0.005,momentum=0.9),loss='mse',metrics=['mae'])
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32, validation_split=0.2,verbose=1)
```

```python
y_pred=model.predict(x_test)
```

```python
print("MSE: {:.4f}".format(mean_squared_error(y_pred,y_test)))
```
